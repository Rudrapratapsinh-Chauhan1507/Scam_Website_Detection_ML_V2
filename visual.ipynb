```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('./dataset/Scam_website_dataset.csv')
```

```python
print("Dataset shape:", df.shape)
```

```python
df.head()
```

## 🔍 Label Distribution

```python
df['label'].value_counts().plot(kind='bar')
plt.title("Label Distribution (0 = Legit, 1 = Scam)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()
```

## 🔗 URL Length Distribution

```python
plt.hist(df['url_length'].dropna(), bins=30)
plt.title("URL Length Distribution")
plt.xlabel("URL Length")
plt.ylabel("Frequency")
plt.show()
```

## ⚠️ Scam Keyword Count

```python
plt.hist(df['scam_keyword_count'].dropna(), bins=30)
plt.title("Scam Keyword Count Distribution")
plt.xlabel("Keyword Count")
plt.ylabel("Frequency")
plt.show()
```

## 🔐 HTTPS vs Label

```python
pd.crosstab(df['https'], df['label']).plot(kind='bar')
plt.title("HTTPS vs Label")
plt.xlabel("HTTPS (1=Yes)")
plt.ylabel("Count")
plt.show()
```

## 🌐 Suspicious TLD vs Label

```python
pd.crosstab(df['suspicious_tld'], df['label']).plot(kind='bar')
plt.title("Suspicious TLD vs Label")
plt.xlabel("Suspicious TLD")
plt.ylabel("Count")
plt.show()
```

## 📝 Text Length Distribution

```python
plt.hist(df['text_length'].dropna(), bins=30)
plt.title("Text Length Distribution")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()
```

## 📈 Feature Correlation (Top Numeric)

```python
corr = df.select_dtypes(include=['int64','float64']).corr()

plt.imshow(corr)
plt.title("Correlation Matrix")
plt.colorbar()
plt.show()
```
